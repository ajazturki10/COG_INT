```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
url = 'https://raw.githubusercontent.com/a-forty-two/COG_GN22CDBDS001_MARCH_22/main/student-mat.csv'

df = pd.read_csv(url)
df.head()
```

```python
df.columns
```

```python
df = df.loc[:, 'school':'guardian']
```

```python
df.head()
```

```python
capitalizer = lambda x : x.capitalize()
```

```python
df['Mjob'] = df['Mjob'].apply(capitalizer)
df['Fjob'] = df['Fjob'].apply(capitalizer)
```

```python
df.tail()
```

```python
def majority(x):
  if x > 17:
    return True
  else:
    return False
```

```python
df['legal_drinker'] = df['age'].apply(majority)
```
